```python
import pandas as pd
import numpy as np
```

```python
file_1 = open('Corpus/Augment1/a_papaya_new.txt')
```

```python
data = file_1.readlines()
```

```python
data
```

```python
# dataset = {'ques':[data[i] for i in range(len(data)) if i % 2 != 0 ],'ans':[]}
```

```python
for i in range(len(data)):
    if data[i] == '===\n':
        del data[i]
```

```python
# for i in range(len(data)):
#     if i % 2 == 0:
#         print(data[i])
dataset = {'ques':[data[i] for i in range(len(data)) if i % 2 == 0 ],
           'ans':[data[i] for i in range(len(data)) if i % 2 != 0]}
```

```python
# dataset['ans'][0][-1]
for i in range(len(dataset['ans'])):
    dataset['ans'][i] =  dataset['ans'][i].replace("\n","")
    dataset['ques'][i] = dataset['ques'][i].replace("\n","") 
```

```python
for i in range(len(dataset['ans'])):
    dataset['ans'][i] =  dataset['ans'][i].replace("A: ","")

for i in range(len(dataset['ques'])):
    dataset['ques'][i] = dataset['ques'][i].replace("Q: ","") 
```

```python
dataset
```

```python
df = pd.DataFrame(dataset)
```
