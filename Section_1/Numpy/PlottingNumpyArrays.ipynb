```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([2,4,5,7,8,9,2,4,6,12])
y = np.random.randint(100,500,10)
```

```python
x,y
```

```python
# by default it will build a line plot
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'ro')
plt.show()
```

```python
plt.scatter(x,y, marker='+',s=120)
plt.show()
```
