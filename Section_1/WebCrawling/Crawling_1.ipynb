```python
import urllib.request as url
import bs4
import pandas as pd
```

```python
page = url.urlopen("http://stats.espncricinfo.com/ci/engine/records/team/largest_margins.html?class=2;id=6;type=team")
```

```python
# lxml - library xml
soup = bs4.BeautifulSoup(page, 'lxml')
```

```python
tables = soup.find_all('table', class_ = "engineTable")
```

```python
# for data in tables:
#     print(data.text)

data = []
for d in tables:
    data.append(d.text)
```

```python
# print(data[0])
data_1 = data[0].replace("\n","\t")
```

```python
data_1 = data_1.replace("\t\t","\t")
```

```python
data_1 = data_1.replace("\t\t\t","\t")
```

```python
data_1 = data_1.replace("\t\t","\t")
```

```python
data_1 = data_1.replace("\t", ",")
```

```python
data_1 = data_1.split("India")
```

```python
df = []
for i in data_1:
    df.append([i])
```

```python
dataset = []
for i in range(len(df)):
    dataset.append(df[i][0].split(","))
```

```python
# print(dataset)
```

```python
df = pd.DataFrame(dataset)
```

```python
df.columns = dataset[0]
```
